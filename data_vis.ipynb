```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("D:\Cancer_Project\Team8_Cancer_ML\HNSCC\HNSCC Clinical Data (Encoded).csv")
```

```python
df.head()
```

```python
corr = df.corr()
print(corr.head())
```

```python
plt.pcolor(corr)
plt.yticks(np.arange(0.5, len(df.columns), 1), df.columns)
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns,rotation='vertical')
plt.margins(0.2)
plt.title("Correlation Graph of HNSCC Dataset")
plt.show()
```

```python
df=pd.read_csv("D:\Cancer_Project\Team8_Cancer_ML\HNSCC-HN1\Cancer Model Performance Data - Prediction (HNSCC-HN1)(CNN).csv")
```

```python
df.head()
```

```python
x=df['Y test']
y=df['Prediction']
```
