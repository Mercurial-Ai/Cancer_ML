```python
df = pd.read_csv("D:\Cancer_Project\Team8_Cancer_ML\METABRIC_RNA_Mutation\Cancer Model Performance Data - Prediction (METABRIC).csv")
```

```python
df.head()
```

```python
x=df["Y_test"]
y=df["Prediction "]
```

```python
plt.style.use("ggplot")
plt.plot(x,y)
plt.xlabel("Y test")
plt.ylabel("Prediction ('chemotherapy')",fontsize=10)
plt.title("Performance of Model on METABRIC Dataset",fontsize=10)
plt.savefig("D:\\Cancer_Project\\Team8_Cancer_ML\\result_graphs\\Pred vs Y_test (METABRIC).png")
plt.show()
```

```python
y = df["Rounded Pred"]
```

```python
plt.style.use("ggplot")
plt.plot(x,y)
plt.xlabel("Y test")
plt.ylabel("Rounded Pred ('chemotherapy')",fontsize=10)
plt.title("Performance of Model on METABRIC Dataset",fontsize=10)
plt.savefig("D:\\Cancer_Project\\Team8_Cancer_ML\\result_graphs\\Rounded Pred vs Y_test (METABRIC).png")
plt.show()
```

```python
# make scatter plots for METABRIC 
y = df['Prediction ']
```

```python
plt.style.use("ggplot")
plt.scatter(x,y)
plt.xlabel("Y test")
plt.ylabel("Prediction ('chemotherapy')",fontsize=10)
plt.title("Performance of Model on METABRIC Dataset",fontsize=10)
plt.savefig("D:\\Cancer_Project\\Team8_Cancer_ML\\result_graphs\\Pred vs Y_test (METABRIC)(scatter).png")
```

```python
# get all coors into one list 
coors = []
for xs in x: 
    for ys in y: 
        point = (xs,ys)
        coors.append(point)
```

```python
# get the num of occurrences for each coor
coorDict = {}
# set a dict loc for each coor
for coor in coors: 
    coorDict[coor]=0
    
for coor in coors: 
    coorDict[coor]=coorDict[coor]+1
    
print(coorDict)
```

```python
coors=coorDict.keys()
x = []
y = []
s = []
for coor in coors: 
    xs = coor[0]
    ys = coor[-1]
    x.append(xs)
    y.append(ys)
    numOc = coorDict[coor]
    s.append(numOc)
```

```python
plt.style.use("ggplot")
plt.scatter(x,y,s=s)
plt.xlabel("Y test")
plt.ylabel("Prediction ('chemotherapy')",fontsize=10)
plt.title("Performance of Model on METABRIC Dataset",fontsize=10)
plt.savefig("D:\\Cancer_Project\\Team8_Cancer_ML\\result_graphs\\Pred vs Y_test (METABRIC)(scatter)(heated).png")
```
