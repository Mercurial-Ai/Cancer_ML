```python
# make scatter plots for METABRIC 
y = df['Prediction ']
```

```python
plt.style.use("ggplot")
plt.scatter(x,y)
plt.xlabel("Y test")
plt.ylabel("Prediction ('chemotherapy')",fontsize=10)
plt.title("Performance of Model on METABRIC Dataset",fontsize=10)
plt.savefig("D:\\Cancer_Project\\Team8_Cancer_ML\\result_graphs\\Pred vs Y_test (METABRIC)(scatter).png")
```

```python
# get all coors into one list 
coors = []
for xs in x: 
    for ys in y: 
        point = (xs,ys)
        coors.append(point)
```

```python
# get the num of occurrences for each coor
coorDict = {}
# set a dict loc for each coor
for coor in coors: 
    coorDict[coor]=0
    
for coor in coors: 
    coorDict[coor]=coorDict[coor]+1
    
print(coorDict)
```

```python
coors=coorDict.keys()
x = []
y = []
s = []
for coor in coors: 
    xs = coor[0]
    ys = coor[-1]
    x.append(xs)
    y.append(ys)
    numOc = coorDict[coor]
    s.append(numOc)
```

```python
plt.style.use("ggplot")
plt.scatter(x,y,s=s)
plt.xlabel("Y test")
plt.ylabel("Prediction ('chemotherapy')",fontsize=10)
plt.title("Performance of Model on METABRIC Dataset",fontsize=10)
plt.savefig("D:\\Cancer_Project\\Team8_Cancer_ML\\result_graphs\\Pred vs Y_test (METABRIC)(scatter)(heated).png")
```

```python
df = pd.read_csv("annual-number-of-deaths-by-cause.csv")
```

```python
df.head()
```

```python
print(df.columns[10:27])
```

```python
cancerCol = df['Deaths - Neoplasms - Sex: Both - Age: All Ages (Number)']
entityCol = df["Entity"]
yearCol = df["Year"]
```

```python
df = pd.DataFrame()
```

```python
df["Entity"]=entityCol
df["Year"]=yearCol
df["Cancers"]=cancerCol
```

```python
df
```

```python
df = df.drop("Entity",axis=1)
```

```python
df
```

```python
df = df.set_index("Year")
df = df.dropna()
smDf=pd.DataFrame()

yearList = []
yearNetList = []
for i in range(11): 
    yearInstance = df.loc[i+2007]
    yearList.append(i+2007)

    yearNet = 0
    for cData in yearInstance["Cancers"]: 
        yearNet = yearNet+cData
    yearNetList.append(yearNet)
        
print(len(yearList))
print(len(yearNetList))
    
smDf["Year"] = yearList
smDf["Cancer"] = yearNetList
    
```

```python
smDf
```

```python
x=smDf["Year"]
y=smDf["Cancer"]
```
