```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# (start, stop, step)
x = np.arange(0, 2*np.pi, 0.1) 
y = np.sin(x)

plt.plot(x, y)
plt.savefig('sin.png', transparent=True)
plt.show()
```

```python
# (start, stop, step)
x = np.arange(0, 2*np.pi, 0.1)
y = np.cos(x)

plt.plot(x, y)
plt.savefig('cos.png', transparent=True)
plt.show()
```
