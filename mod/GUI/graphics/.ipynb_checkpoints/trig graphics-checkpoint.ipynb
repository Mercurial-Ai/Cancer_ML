```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
for i in range(50): 

    # (start, stop, step)
    x = np.arange(0, 2*np.pi+i, 0.1) 
    y = np.sin(x)

    plt.plot(x, y)

    # (start, stop, step)
    x = np.arange(0, 2*np.pi+i, 0.1)
    y = np.cos(x)

    plt.plot(x, y)
    plt.savefig('graph'+str(i)+'.png', transparent=True)
    plt.show()
```
